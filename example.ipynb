```python
import numpy as np
import numpy.random as random
import matplotlib.pyplot as plt

from kd_tree.kd_tree import KDTree
```

```python
rgen = random.RandomState(9)
points = rgen.rand(1000, 2)
```

```python
tree = KDTree(points)
print(tree.query_knn([1., 1.], 3))
```

```python
tree.visualize(points)
plt.show()
```
