```python
import numpy as np
import numpy.random as random
import matplotlib.pyplot as plt
from sklearn import datasets

from kd_tree.kd_tree import KDTree
```

```python
plt.rc("figure", autolayout=True)
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=14,
    titlepad=10,
)
%matplotlib inline
```

```python
rgen = random.RandomState(9)
points = rgen.rand(1000, 2)
```

```python
tree = KDTree(points)
print(tree.query_knn([1., 1.], 3, sort=True))
```

```python
ax = tree.visualize()
ax.set_title("kd-tree of 1000 uniformly distributed points")
plt.show()
```

```python
iris = datasets.load_iris()
X = iris.data[:, [0, 3]]
iris_tree = KDTree(X)
ax = iris_tree.visualize()
ax.set_title("kd-tree of iris dataset")
plt.show()
```

```python
tree = KDTree([1, 1])
print(tree.query_knn([0.3, 0.3], 1, sort=True))
tree.add_point(np.array([0, 0]))
print(tree.query_knn([0.3, 0.3], 1, sort=True))
tree.add_point(np.array([0.1, 0.1]))
print(tree.query_knn([0.3, 0.3], 1, sort=True))
tree.add_point(np.array([0.5, 0.4]))
print(tree.query_knn([0.3, 0.3], 3, sort=True, return_distances=True))
ax = tree.visualize()
```

```python
tree = KDTree([1, 1])
tree.add_point([0, 0])
tree.add_point([0.1, 0.1])
tree.add_point([0.5, 0.4])
tree.add_point([0.3, 0.3])
tree.add_point([0.6, 0.6])
ax = tree.visualize()
```

```python
X = np.array([[1, 1],
              [0, 0],
              [0.1, 0.1],
              [0.5, 0.4]])
tree = KDTree(X)
tree.visualize()
```
